```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import xarray as xr
%matplotlib inline
#matplotlib.rcParams['axes.labelsize'] = 18
matplotlib.rcParams['xtick.labelsize'] = 20
matplotlib.rcParams['ytick.labelsize'] = 20
matplotlib.rcParams['text.color'] = 'k'
matplotlib.rcParams['figure.figsize'] = 20,20
```

# Urban1

```python
lon1, lat1 = (278641, 2106131)    
xarr1_0 = xr.open_rasterio(r'D:\pixxel\EO1H1480472016328110PZ_1GST\tile2.tiff')
img1_0 = xarr1_0[:, :, :] 
val1_0 = img1_0.sel(x=lon1, y=lat1, method="nearest")    
print("This is val: ", val1_0)
```

```python
xarr1_1 = xr.open_rasterio(r'D:\pixxel\EO1H1480472016328110PZ_1GST\tile2_rb.tiff')
img1_1 = xarr1_1[:, :, :] 
val1_1 = img1_1.sel(x=lon1, y=lat1, method="nearest")    
print("This is val: ", val1_1)
```

```python
xarr1_2 = xr.open_rasterio(r'D:\pixxel\EO1H1480472016328110PZ_1GST\quac_new\rad.tif')
img1_2 = xarr1_2[:, :, :] 
val1_2 = img1_2.sel(x=lon1, y=lat1, method="nearest")    
print("This is val: ", val1_2)
```

```python
xarr1_3 = xr.open_rasterio(r'D:\pixxel\EO1H1480472016328110PZ_1GST\QUAC\QUAC_ready.tif')
img1_3 = xarr1_3[:, :, :] 
val1_3 = img1_3.sel(x=lon1, y=lat1, method="nearest")    
print("This is val: ", val1_3)
```

```python
xarr1_4 = xr.open_rasterio(r'C:\Users\Sumana\Pixxel\TASK\LEVEL3_SAVGOL_TIFF\tile2_hfn.tiff')
img1_4 = xarr1_4[:, :, :] 
val1_4 = img1_4.sel(x=lon1, y=lat1, method="nearest")    
print("This is val: ", val1_4)
```

```python
fig, axs = plt.subplots(5)
fig.suptitle('Spectral profile of Urban1',fontsize=22, fontweight='bold')
axs[0].plot(val1_0)
axs[0].set_title('Actual data',fontsize=18)
axs[1].plot(val1_1)
axs[1].set_title('Actual data after Bad Band removal',fontsize=18)
axs[2].plot(val1_2)
axs[2].set_title('Radiance data',fontsize=18)
axs[3].plot(val1_3/10000)
axs[3].set_title('QUAC Reflectance data',fontsize=18)
axs[4].plot(val1_4/10000)
axs[4].set_title('Smoothened QUAC Reflectance data (Savgol Filter)',fontsize=18)
fig. tight_layout(pad=6.0)
```
